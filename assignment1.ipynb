## 📋 Data Cleaning Summary

**Dataset:** Combined Shopping Dataset   
**Source:** Combined_dataset.csv  

### Shape
- **Original Shape:** 1000 rows × 24 columns  
- **Final Shape:** Check using `df.shape` after cleaning  

### Missing Values
- Identified missing values using `df.isnull().sum()`  
- Filled missing `discount` values with **mean**  
- Dropped rows with missing values in critical columns  

### Duplicates
- Checked duplicates using `df.duplicated().sum()`  
- Removed all duplicate rows using `drop_duplicates()`  

### Derived Column
- Created new column: **`total_amount = final_price × quantity`**  
- `quantity` column was added with default value of 1  

### Output
- Cleaned dataset saved as **`cleaned_dataset.csv`**  
- Key columns used: `product_id`, `title`, `rating`, `initial_price`, `discount`, `final_price`, `category`, `total_amount`
